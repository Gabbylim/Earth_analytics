```python
#subsetting data
# 1)spatial
lat = 39.98
lon=240.0

# 2) temporal
start_time = "1950-01-01"
end_time = "2005-12-15"
latitude,longitude=min_temp_xr.lat.values[1],min_temp_xr.lon.values[1]
latitude,longitude


```

```python
#slicing
min_temp_single_point = min_temp_xr["air_temperature"].sel(lat=latitude,lon=longitude)
min_temp_single_point
```

```python
#clean up plot
f,ax = plt.subplots(figsize=(12,6))
temp_single_point.plot.line(hue='lat',marker="o",ax=ax,color="grey",markerfacecolor="purple",markeredgecolor="purple")
ax.set(title="Time series for a single point")
plt.show()
```

```python
#converting climate time series data to a pandas dataframe
temp_single_point_df = temp_single_point.to_dataframe()
temp_single_point_df
```

```python
#temporal slicing  of netcdf4 data
temp_1950_2005 = min_temp_xr['air_temperature'].sel(time = slice(start_time,end_time),lat=latitude,lon=longitude)
temp_1950_2005.shape
#plotting data
f,ax = plt.subplots(figsize=(12,6))
temp_1950_2005.plot.line(hue='lat',marker='*',ax=ax,color="black",markerfacecolor="purple",markeredgecolor="purple")
ax.set(title="5-year time series of temp data for a specific time period")
plt.show()

```

```python
#exporting  data
min_temp_df = temp_1950_2005.to_dataframe()
min_temp_df.head()
file_name = "min_monthly_temp-" + start_time + "-" + end_time + ".csv"
min_temp_df.to_csv(file_name)
```
